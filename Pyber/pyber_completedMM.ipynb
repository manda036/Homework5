```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# city_data.head()
# ride_data.head()

# Combine the data into a single dataset
merged_data = pd.merge(ride_data, city_data, how="outer")

# Display the data table for preview
merged_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_data = merged_data.loc[merged_data["type"]=="Urban"]
urban_count = urban_data.groupby("city").count()
urban_xaxis= urban_count["ride_id"]
urban_average = urban_data.groupby("city").mean()
urban_yaxis = urban_average["fare"]
urban_size = urban_data.groupby("city")
urban_circle_size = urban_size["driver_count"].mean()


rural_data = merged_data.loc[merged_data["type"]=="Rural"]
rural_count = rural_data.groupby("city").count()
rural_xaxis= rural_count["ride_id"]
rural_average = rural_data.groupby("city").mean()
rural_yaxis = rural_average["fare"]
rural_size = rural_data.groupby("city")
rural_circle_size = rural_size["driver_count"].mean()

suburban_data = merged_data.loc[merged_data["type"]=="Suburban"]
suburban_count = suburban_data.groupby("city").count()
suburban_xaxis= suburban_count["ride_id"]
suburban_average = suburban_data.groupby("city").mean()
suburban_yaxis = suburban_average["fare"]
suburban_size = suburban_data.groupby("city")
suburban_circle_size = suburban_size["driver_count"].mean()

labels_list = ["Urban", "Suburban", "Rural"]

# Build the scatter plots for each city types
urban = plt.scatter(urban_xaxis, urban_yaxis, marker="o", s=urban_circle_size*10, facecolor="lightcoral", edgecolor="black", alpha=.75)
suburban = plt.scatter(suburban_xaxis, suburban_yaxis, marker="o", s=suburban_circle_size*10, facecolor="lightskyblue", edgecolor="black", alpha=.75)
rural = plt.scatter(rural_xaxis, rural_yaxis, marker="o", s=rural_circle_size*10, facecolor="gold", edgecolor="black", alpha=.75)

plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (per City)")
plt.ylabel("Average Fare ($)")

# Incorporate the other graph properties
plt.grid()

# Create a legend
plt.legend(title="City Types", labels=labels_list, loc="best")

# Incorporate a text label regarding circle size
plt.text(42, 35, "Note: Circle size correlates to driver count per city")

# Save Figure
plt.savefig("Pyber_Scatter_Plot.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urbanfare = urban_data["fare"].sum()
suburbanfare = suburban_data["fare"].sum()
ruralfare = rural_data["fare"].sum()
percents= [urbanfare, suburbanfare, ruralfare]

# Build Pie Chart
labels = ["Urban", "Suburban", "Rural"]
colors = ["lightcoral", "lightskyblue", "gold"]
explode = [.1, 0, 0]

plt.pie(percents, labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", shadow=True, startangle=270)
plt.axis("equal")
plt.title("% of Total Fare by City Type")

# Save Figure
plt.savefig("Total_Fare_Pie_Chart.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urbanrides = urban_data["ride_id"].count()
suburbanrides = suburban_data["ride_id"].count()
ruralrides = rural_data["ride_id"].count()
percents2= [urbanrides, suburbanrides, ruralrides]

# Build Pie Chart
labels = ["Urban", "Suburban", "Rural"]
colors = ["lightcoral", "lightskyblue", "gold"]
explode = [.1, 0, 0]

plt.pie(percents2, labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", shadow=True, startangle=270)
plt.axis("equal")
plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("Total_Rides_Pie_Chart.png")
```

```python
# Show Figure
plt.show()
```
